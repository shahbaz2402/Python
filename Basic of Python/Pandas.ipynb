```python
stu_detls = {'s_id':[1,2,3,4,5],
           'stu_name':['A','B','C','D','E'],
           'Region':['N','S','E','E','E']}

marks_detls = {'stu_id':[1,2,3,4,6],
             'sub_name':['Py','R','Tab','Sql','Py'],
              'Marks':[32,22,32,33,35,]}

stu_detls
      
```
