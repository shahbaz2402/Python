```python
sales =[4522,5443,5466,4363,5566]
profit =[343,234,344,654,453]
year =[2016,2017,2018,2019,2020]
```

```python
plt.plot(year,sales, 'Black')
plt.xlabel('year')
plt.ylabel('sales and profit')
plt.title('Sales and profit by year')
plt.plot(year,profit, 'Red')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(year,sales, 'Blue')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('Sales by year')

plt.subplot(2,1,2)
plt.plot(year,profit, 'red')
plt.xlabel('year')
plt.ylabel('profit')
plt.title('profit by year')
plt.tight_layout()
```

```python
plt.subplot(1,2,1)
plt.plot(year,sales, 'Blue')
plt.xlabel('year')
plt.ylabel('Sales')
plt.title('Sales by year')

plt.subplot(1,2,2)
plt.plot(year,profit, 'red')
plt.xlabel('year')
plt.ylabel('Profit')
plt.title('Profit by year')
plt.tight_layout()
```
