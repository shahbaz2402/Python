```python
(flight_data >>
select (X.origin, X.dest, X.hour))
```

```python
(flight_data >>
select (~X.origin, ~X.dest, ~X.hour))

#(~ is used for except this origin, dest, hour data)
```

```python
(flight_data >>
mask(X.month==1, X.day==1, X.origin=='JFK',X.hour>10))
```

```python
flight_data
```

```python
(flight_data >>
arrange(X.distance,X.hour,ascending=True))
```

```python
#mutate is use to create new column
(flight_data >>
mutate(new_dist = X.distance/1000,carrier_origin =X.carrier + X.origin))
 
```

```python
flight_data.origin.unique()
```

```python
(flight_data >>
group_by (X.origin) >>
summarise(mean_dist = X.distance.mean()))
```

```python
(flight_data >>
mask(X.hour >10) >>
mutate(speed =X.distance/(X.air_time*60))>>
group_by (X.origin)>>
summarise(mean_speed =X.speed.mean())>>
arrange(X.mean_speed,ascending = False))
```
